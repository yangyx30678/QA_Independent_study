Load

```python
import os
import json
import numpy
dataset_dir = '../datasets/zhwiki_odqa'

for sp in os.listdir(dataset_dir):
    data = []
    print(sp)
    for file_dir in os.listdir(f'{dataset_dir}/{sp}'):
        with open(f'{dataset_dir}/{sp}/{file_dir}', 'r', encoding='utf8')  as file:
            data_batch = [json.loads(each) for each in file.readlines()]
            data.extend([{'id': each['id'], 'title':each['title'], 'text':each['text']} for each in data_batch])
    numpy.save(f'./data_od/{sp}', data)
print()
```

整理分詞

```python
import os
import jieba
import numpy

data_file = './data_od/'
data = []

for npyfile in os.listdir(data_file):
    kwds = []
    data = numpy.load(data_file+npyfile, allow_pickle=True)
    for i in range(len(data)):
        for kw in jieba.cut(data[i]['text']):
            if kw not in kwds:
                kwds.append(kw)
        if i % 10 == 0:
            print(f'{npyfile}/{i}/{len(data)}')
        if i % 5000 == 0: 
            numpy.save(f'./data4finetuning/dict/{npyfile[:2]}{i}', kwds)
            kwds = []
    numpy.save(f'./data4finetuning/dict/{npyfile[:2]}', kwds)
```
